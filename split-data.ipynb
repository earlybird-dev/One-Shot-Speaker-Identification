```python
import os
import re

from scipy import signal
import soundfile as sf
import math

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sort a file list
# https://stackoverflow.com/a/48030307
def sorted_alphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)
```

```python
# Reorganize spectrograms into folders
spectrograms = sorted_alphanumeric(os.listdir('./spectrogram/'))
spectrograms
```

```python
for i in range(len(spectrograms)):
    person =  '%02d' % math.floor(i/4)
    try:
        os.mkdir(f"./spectrogram/train/")
        os.mkdir(f"./spectrogram/test/")
    except OSError as error: 
        pass
    
    try:
        os.mkdir(f"./spectrogram/train/{person}")
        os.mkdir(f"./spectrogram/test/{person}")
    except OSError as error: 
        pass
    
    if i%4!=3:
        os.rename(f"./spectrogram/{spectrograms[i]}", f"./spectrogram/train/{person}/{spectrograms[i]}")
    else:
        os.rename(f"./spectrogram/{spectrograms[i]}", f"./spectrogram/test/{person}/{spectrograms[i]}")
```
